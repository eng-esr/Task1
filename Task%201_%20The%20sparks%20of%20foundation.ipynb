```python
q1 = store.describe()['Profit']['25%']
q3 = store.describe()['Profit']['75%']
iqr = q3 - q1
max_profit = q3 + 1.5 * iqr
print(max_profit)
```

```python
outliers = store[store['Profit'] >= max_profit]
outliers_count = outliers['Profit'].count()
store_count = store['Profit'].count()
print('Percentage removed: ' + str(round(outliers_count/store_count * 100, 2)) + '%')
```
